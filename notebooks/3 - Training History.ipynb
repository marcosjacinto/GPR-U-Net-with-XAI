```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
```

```python
source_path = '../model/models/'

fname1 = 'model_5x5_8filters_complete_history.pkl'
fname2 = 'model_5x5_8filters_only_0_history.pkl'
fname3 = 'model_5x5_8filters_only_1_history.pkl'
fname4 = 'model_5x5_8filters_only_2_history.pkl'
fname5 = 'model_5x5_8filters_only_3_history.pkl'
fname6 = 'model_5x5_8filters_only_4_history.pkl'
fname7 = 'model_5x5_8filters_only_5_history.pkl'
```

```python
def process_df(fname):

    with open(source_path + fname, 'rb') as file:   
        history = pickle.load(file)

    df = pd.DataFrame(history)
  

    try:
        df.drop(columns = ['dice', 'val_dice'], inplace = True)
    except:
        pass

    df['F1 Score'] = 2 * (df['precision'] * df['recall']) / (df['precision'] + df['recall'])
    df['Val. F1 Score'] = 2 * (df['val_precision'] * df['val_recall']) / (df['val_precision'] + df['val_recall'])

    df.columns = ['Training loss function',
    'tp',
    'fp',
    'tn',
    'fn',
    'Training accuracy',
    'Training precision',
    'Training recall',
    'Training AUC',
    'Validation loss function',
    'val_tp',
    'val_fp',
    'val_tn',
    'val_fn',
    'Validation accuracy',
    'Validation precision',
    'Validation recall',
    'Validation AUC',
    'Training F1 Score',
    'Validation F1 Score']

    return df
```

```python
metrics = ['Training accuracy', 'Training precision', 'Training AUC']
val_metrics = ['Validation accuracy', 'Validation precision', 'Validation AUC']
size = '12'
colors = {
    'Training accuracy' : 'black', 'Training precision' : 'red', 'Training AUC' : 'blue',
    'Validation accuracy' : 'black', 'Validation precision' : 'red', 'Validation AUC' : 'blue',
}

outputPath = '../model/history/'

fig, ax = plt.subplots(4, 2, figsize = (10, 20))

ax[-1, -1].remove()

i, j = 0, 0
df = process_df(fname1)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with six features')

########

i, j = 0, 1
df = process_df(fname2)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with GPR Section')

########

i, j = 1, 0
df = process_df(fname3)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with Similarity')

########

i, j = 1, 1
df = process_df(fname4)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with Energy')

########

i, j = 2, 0
df = process_df(fname5)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with Instantaneous Frequency')

########

i, j = 2, 1
df = process_df(fname6)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with Instantaneous Phase')

########

i, j = 3, 0
df = process_df(fname7)
for metric in metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'solid', label = metric)

for metric in val_metrics:
    ax[i, j].plot(df[metric], color = colors[metric], linestyle = 'dashed', label = metric)

ax[i, j].legend(fontsize=size)
ax[i, j].set_xlabel('Epochs')
ax[i, j].set_ylabel('Evaluation metrics')
ax[i, j].set_title('Model with Hilbert Trace / Similarity')

fig.tight_layout()
fig.savefig(f'{outputPath}metrics.png', dpi = 300)
```
