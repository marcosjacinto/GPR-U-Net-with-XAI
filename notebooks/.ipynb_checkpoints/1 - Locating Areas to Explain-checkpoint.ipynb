```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
data_path = '../data/processed/'

x_test = np.load(data_path + 'x_test.npy')
y_test = np.load(data_path + 'y_test.npy')
x_test_section_2 = np.load(data_path + 'x_test_section_96.npy')
y_test_section_2 = np.load(data_path + 'y_test_section_96.npy')
y_test_section_2 = y_test_section_2[:928, :]
```

## Exploring possible areas that will be explained from the test section 96

```python
n = 0
sampleSize = 16
step_div = 4
step = int(sampleSize/step_div)
shape1, shape2 = y_test_section_2.shape[0], y_test_section_2.shape[1]

for i in range(int(shape1/(step)) - (step_div - 1)):
  istart = int(step * i)
  iend = int(istart + sampleSize)
  for w in range(int(shape2/(step)) - (step_div - 1)):
    wstart = int(step * w)
    wend = int(wstart + sampleSize)
    if np.array_equal(y_test_section_2[istart : iend, wstart : wend], y_test[7882]):
      print(istart, iend, wstart, wend)
```
